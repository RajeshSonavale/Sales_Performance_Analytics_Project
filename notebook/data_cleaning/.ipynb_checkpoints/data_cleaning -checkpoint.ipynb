# Data Modeling 

### Step 1 : Load clean data

```python
df=pd.read_csv('C:/data_science/Projects/cloud_sales_analysis/data/cleaned/sales_cleaned.csv')
```

```python
df.head()
```

### Step 2 : Create a dimension table

```python
# dim_customer
dim_customer = df[['customer_id', 'customer_name', 'segment']].drop_duplicates()
dim_customer.head()
```

```python
# dim_product 
dim_product = df[['product_id', 'product_name', 'category', 'sub_category']].drop_duplicates()
dim_product
```

```python
# dim_date
dim_date = df[['order_date']].drop_duplicates()

dim_date['year'] = pd.to_datetime(dim_date['order_date']).dt.year
dim_date['month'] = pd.to_datetime(dim_date['order_date']).dt.month
dim_date['quarter'] = pd.to_datetime(dim_date['order_date']).dt.quarter
dim_date
```

### Step 3 : Create fact table

```python
# fact_sales
fact_sales = df[['order_id', 'product_id', 'customer_id',
                 'order_date', 'ship_date',
                 'sales', 'quantity', 'discount', 'profit']]
fact_sales
```

```python
dim_customer.to_csv('C:/data_science/Projects/cloud_sales_analysis/data/cleaned/dim_customer.csv', index=False)
dim_product.to_csv('C:/data_science/Projects/cloud_sales_analysis/data/cleaned/dim_product.csv', index=False)
dim_date.to_csv('C:/data_science/Projects/cloud_sales_analysis/data/cleaned/dim_date.csv', index=False)
fact_sales.to_csv('C:/data_science/Projects/cloud_sales_analysis/data/cleaned/fact_sales.csv', index=False)
```

### Step 4 : import data in Mysql

```python
import pandas as pd
from sqlalchemy import create_engine

# Step 1: Load CSV
df = pd.read_csv('C:/data_science/Projects/cloud_sales_analysis/data/cleaned/dim_product.csv')

# Step 2: Clean column names (important)
df.columns = df.columns.str.strip()

# Step 3: MySQL connection (password encoded)
engine = create_engine("mysql+pymysql://root:Rajesh%40123@localhost:3306/sales_project")

# Step 4: Load into MySQL
df.to_sql(name="dim_product", con=engine, if_exists="replace", index=False)

print("Data loaded successfully 🚀")
```
